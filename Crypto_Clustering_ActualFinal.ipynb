```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import holoviews as hv
from holoviews import opts
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_data = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']])
scaled_data_df= pd.DataFrame(scaled_data, columns=['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'], index=df_market_data.index)
```

```python

# Display sample data
scaled_data_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model = KMeans(n_clusters=k, random_state=0)
    # 2. Fit the model to the data using `df_market_data_scaled`
    kmeans_model.fit(scaled_data_df)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": list(range(1, 12)), "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
elbow_plot = df_elbow.hvplot.line(x='k', y='inertia', title='Elbow Curve', xticks=k_values, xlabel='Number of Clusters (k)', ylabel='Inertia')

# the different values of k to visually identify the optimal value for k.
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k' is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
kmeans_model= KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(scaled_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_predictions = kmeans_model.predict(scaled_data_df)

# Print the resulting array of cluster values.
kmeans_predictions
```

```python
# Create a copy of the DataFrame
kmeanspredictions_df = scaled_data_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
kmeanspredictions_df["Clusters"] = kmeans_predictions
# Display sample data
kmeanspredictions_df
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
kmeanspredictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Clusters",
    hover_cols=["coin_id"]
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
pca_features = pca_model.fit_transform(scaled_data_df)
# three principal components.
df_pca = pd.DataFrame(
    data=pca_features, 
    columns=['principal_component_1', 'principal_component_2', 'principal_component_3'], 
    index=scaled_data_df.index
)
# View the first five rows of the DataFrame. 
df_pca.head(5)
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
variance_ratio = pca_model.explained_variance_ratio_
print(variance_ratio)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total variance is .88 or 88%

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(
    data=pca_features, 
    columns=['principal_component_1', 'principal_component_2', 'principal_component_3'], 
    index=scaled_data_df.index
)


# Display sample data
df_pca.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values2 = list(range(1, 12))
```

```python
# Create an empy list to store the inertia values
inertia2 = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values2:
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans_model2 = KMeans(n_clusters=k, random_state=0)
    # 2. Fit the model to the data using `df_market_data_pca`
    kmeans_model2.fit(df_pca)
    # 3. Append the model.inertia_ to the inertia list
    inertia2.append(kmeans_model2.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data2 = {"k": k_values2, "inertia": inertia2}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow2 = pd.DataFrame(elbow_data2)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow2.hvplot.line(x="k", y="inertia", title="Elbow Curve with PCA Data")
```

  * **Answer:** The best value for K is 4

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
kmeans_model2= KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
kmeans_model2.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
kmeans_predictions2 = kmeans_model2.predict(df_pca)
# Print the resulting array of cluster values.
kmeans_predictions2
```

```python
# Create a copy of the DataFrame with the PCA data
pca_predicted_df = df_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_predicted_df["Clusters"] = kmeans_predictions2

# Display sample data
pca_predicted_df
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_predicted_df.hvplot.scatter(
    x="principal_component_1",
    y="principal_component_2",
    by="Clusters",
    hover_cols=["coin_id"]
)
```
